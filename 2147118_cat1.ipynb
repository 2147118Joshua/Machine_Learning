```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn.cluster import AgglomerativeClustering
import pandas as pd
```

```python
data_set = pd.read_csv("segmented_customers.csv.csv")
```

```python
print(data_set.head())
```

```python
data_set.describe()
```

```python
#Extracting Independent and dependent Variable  
x= data_set.iloc[:10, 3].values  
y= data_set.iloc[:10,4].values
```

```python
print(x)
print(y)
```

```python
data = list(zip(x, y))
print(data)
```

```python
linkage_data = linkage(data, method='complete', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python
hierarchical_cluster = AgglomerativeClustering(n_clusters=2, affinity='euclidean',
linkage='complete')
```

```python
linkage_data = linkage(data, method='ward', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python

linkage_data = linkage(data, method='average', metric='euclidean')
dendrogram(linkage_data)
plt.show()
```

```python

labels = hierarchical_cluster.fit_predict(data)
print(labels)
plt.scatter(x, y, c=labels)
plt.show()
```
